## 6. Test configuration Architecture Medallion

```python
# === CONFIGURATION INLINE - Architecture Medallion (Bronze/Silver/Gold) ===
from pathlib import Path

# Détection automatique de la racine du projet
notebook_dir = Path.cwd()
if notebook_dir.name == 'notebooks':
    PROJECT_ROOT = notebook_dir.parent
else:
    PROJECT_ROOT = notebook_dir

# Chemins des données - Medallion Architecture
DATA_BRONZE = PROJECT_ROOT / 'data' / 'bronze'  # Données brutes
DATA_SILVER = PROJECT_ROOT / 'data' / 'silver'  # Données nettoyées
DATA_GOLD = PROJECT_ROOT / 'data' / 'gold'      # Données ML-ready

# Paramètres métier
DEPARTEMENTS_PETITE_COURONNE = ['75', '92', '93', '94']
COMMUNES_PETITE_COURONNE_COUNT = 144  # Validé depuis données INSEE 2024

print("📁 Configuration - Architecture Medallion (Bronze/Silver/Gold):")
print(f"🥉 Bronze: {DATA_BRONZE}")
print(f"🥈 Silver: {DATA_SILVER}")
print(f"🥇 Gold: {DATA_GOLD}")
print(f"\n📍 Départements Petite Couronne: {DEPARTEMENTS_PETITE_COURONNE}")
print(f"📊 Nombre de communes: {COMMUNES_PETITE_COURONNE_COUNT}")
print("\n✅ Configuration chargée avec succès")
```

## 7. Vérification structure des dossiers

```python
# Vérification de la structure Medallion
layers = [
    ('Bronze (données brutes)', DATA_BRONZE),
    ('Silver (données nettoyées)', DATA_SILVER),
    ('Gold (données agrégées)', DATA_GOLD)
]

print("📁 Vérification structure Medallion:")
for name, path in layers:
    exists = "✅" if path.exists() else "❌"
    print(f"{exists} {name}: {path.name}")
```

## 8. Test mémoire

Vérification que pandas peut gérer nos datasets

```python
# Vérifier l'utilisation mémoire
import psutil

mem = psutil.virtual_memory()
print(f"Mémoire disponible: {mem.available / (1024**3):.2f} GB")
print(f"Mémoire totale: {mem.total / (1024**3):.2f} GB")
print(f"Utilisation: {mem.percent}%")

if mem.available > 2 * (1024**3):
    print("✅ Mémoire suffisante pour nos datasets (<10GB)")
else:
    print("⚠️ Mémoire limitée - traiter les données par chunks si nécessaire")
```
