## 1️⃣ Configuration et Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Configuration pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)

print("✅ Imports chargés")
```

```python
# Chemins Medallion Architecture
notebook_dir = Path.cwd()
PROJECT_ROOT = notebook_dir.parent if notebook_dir.name == 'notebooks' else notebook_dir

DATA_BRONZE = PROJECT_ROOT / 'data' / 'bronze'
DATA_SILVER = PROJECT_ROOT / 'data' / 'silver'
DATA_GOLD = PROJECT_ROOT / 'data' / 'gold'

# Paramètres
DEPARTEMENTS_PC = ['75', '92', '93', '94']

print(f"🥉 Bronze : {DATA_BRONZE}")
print(f"🎯 Périmètre : {', '.join(DEPARTEMENTS_PC)}")
```

```python
# Charger population départementale (grandes classes d'âge)
df_pop_gca = pd.read_excel(
    DATA_BRONZE / "population_dep_gca_1975_2023.xls",
    engine='xlrd'  # Format ancien Excel .xls
)

print("📊 POPULATION DÉPARTEMENTALE - GRANDES CLASSES D'ÂGE")
print("=" * 60)
print(f"Dimensions : {df_pop_gca.shape[0]:,} lignes × {df_pop_gca.shape[1]} colonnes")
print(f"\nColonnes disponibles :")
print(list(df_pop_gca.columns))
print(f"\nÉchantillon :")
df_pop_gca.head()
```

```python
# Identifier colonnes clés
print("🔍 Types de colonnes :")
print(df_pop_gca.dtypes)

print("\n💡 Valeurs manquantes :")
missing = df_pop_gca.isnull().sum()
if missing.sum() > 0:
    print(missing[missing > 0])
else:
    print("✅ Aucune valeur manquante")
```

```python
# Filtrer Petite Couronne (si colonne département disponible)
# À adapter selon la structure réelle du fichier
print("\n🎯 Tentative de filtrage Petite Couronne...")
print("\n👁️  Premières lignes pour identifier structure :")
print(df_pop_gca.head(10))
```

## 3️⃣ Exploration Naissances 2008-2024

```python
# Lister fichiers dans dossier naissances
naissances_dir = DATA_BRONZE / "naissances_2008_2024"
print("📁 Fichiers dans naissances_2008_2024 :")
for f in naissances_dir.iterdir():
    size_mb = f.stat().st_size / 1e6
    print(f"  - {f.name} ({size_mb:.2f} MB)")
```

```python
# Trouver le fichier CSV principal
csv_files = list(naissances_dir.glob("*.csv"))
if csv_files:
    csv_file = csv_files[0]
    print(f"\n📥 Chargement : {csv_file.name}")
    
    # Charger échantillon pour explorer structure
    df_naissances = pd.read_csv(csv_file, nrows=1000, sep=';')
    
    print("\n📊 NAISSANCES 2008-2024")
    print("=" * 60)
    print(f"Dimensions échantillon : {df_naissances.shape[0]:,} lignes × {df_naissances.shape[1]} colonnes")
    print(f"\nColonnes disponibles :")
    print(list(df_naissances.columns))
    print(f"\nTypes :")
    print(df_naissances.dtypes)
else:
    print("❌ Aucun fichier CSV trouvé")
```

```python
# Afficher échantillon
if csv_files:
    print("\n👁️  Échantillon données :")
    print(df_naissances.head(10))
    
    # Statistiques descriptives
    print("\n📈 Statistiques descriptives :")
    print(df_naissances.describe())
```

## 4️⃣ Exploration Décès 2008-2024

```python
# Lister fichiers dans dossier décès
deces_dir = DATA_BRONZE / "deces_2008_2024"
print("📁 Fichiers dans deces_2008_2024 :")
for f in deces_dir.iterdir():
    size_mb = f.stat().st_size / 1e6
    print(f"  - {f.name} ({size_mb:.2f} MB)")
```

```python
# Trouver le fichier CSV principal
csv_files_deces = list(deces_dir.glob("*.csv"))
if csv_files_deces:
    csv_file_deces = csv_files_deces[0]
    print(f"\n📥 Chargement : {csv_file_deces.name}")
    
    # Charger échantillon
    df_deces = pd.read_csv(csv_file_deces, nrows=1000, sep=';')
    
    print("\n📊 DÉCÈS 2008-2024")
    print("=" * 60)
    print(f"Dimensions échantillon : {df_deces.shape[0]:,} lignes × {df_deces.shape[1]} colonnes")
    print(f"\nColonnes disponibles :")
    print(list(df_deces.columns))
    print(f"\nÉchantillon :")
    print(df_deces.head(10))
else:
    print("❌ Aucun fichier CSV trouvé")
```

## 5️⃣ Exploration Référentiels COG 2023-2024

```python
# Lister référentiels COG disponibles
cog_dir = DATA_BRONZE / "referentiels_cog"
print("📁 Référentiels COG disponibles :")
for f in sorted(cog_dir.glob("*.csv")):
    size_mb = f.stat().st_size / 1e6
    print(f"  - {f.name} ({size_mb:.2f} MB)")
```

```python
# Charger COG 2024 (le plus récent)
cog_2024 = cog_dir / "referentiel_communes_2024.csv"
if cog_2024.exists():
    df_cog_2024 = pd.read_csv(cog_2024)
    
    print("📊 RÉFÉRENTIEL COG 2024")
    print("=" * 60)
    print(f"Dimensions : {df_cog_2024.shape[0]:,} lignes × {df_cog_2024.shape[1]} colonnes")
    print(f"\nColonnes :")
    print(list(df_cog_2024.columns))
    print(f"\nÉchantillon :")
    print(df_cog_2024.head())
else:
    print("❌ COG 2024 non trouvé")
```

```python
# Filtrer Petite Couronne
if cog_2024.exists():
    # Extraire département depuis code commune
    df_cog_2024['dept'] = df_cog_2024['COM'].astype(str).str[:2]
    
    df_cog_pc = df_cog_2024[df_cog_2024['dept'].isin(DEPARTEMENTS_PC)]
    
    print("\n🎯 COMMUNES PETITE COURONNE (COG 2024)")
    print("=" * 60)
    print(f"Total communes : {len(df_cog_pc)}")
    print(f"\nRépartition par département :")
    print(df_cog_pc.groupby('dept').size().sort_index())
    
    print(f"\n💾 Sauvegarde liste codes INSEE Petite Couronne...")
    codes_insee_pc = df_cog_pc['COM'].tolist()
    print(f"✅ {len(codes_insee_pc)} codes INSEE identifiés")
```

## 6️⃣ Analyse des Clés de Jointure

```python
print("🔑 ANALYSE DES CLÉS DE JOINTURE")
print("=" * 60)

# Identifier colonnes géographiques dans chaque dataset
datasets_info = {
    "COG 2024": {
        "df": df_cog_2024 if cog_2024.exists() else None,
        "cols_geo": ['COM', 'DEP', 'dept']
    },
    "Naissances": {
        "df": df_naissances if csv_files else None,
        "cols_geo": [c for c in (df_naissances.columns if csv_files else []) if 'code' in c.lower() or 'geo' in c.lower()]
    },
    "Décès": {
        "df": df_deces if csv_files_deces else None,
        "cols_geo": [c for c in (df_deces.columns if csv_files_deces else []) if 'code' in c.lower() or 'geo' in c.lower()]
    }
}

for name, info in datasets_info.items():
    if info['df'] is not None:
        print(f"\n📊 {name}")
        print(f"   Colonnes géographiques : {info['cols_geo']}")
        if info['cols_geo']:
            print(f"   Échantillon :")
            print(info['df'][info['cols_geo']].head(3))
    else:
        print(f"\n⚠️  {name} : non chargé")
```
